```python
def clean_df(df, max_unique_values=20, verbose=True):
    df_copy = df.copy()
    changed = []

    # cast low-cardinality integer columns to 'category'
    int_dtypes = ["int", "int32", "int64", "Int32", "Int64"]  # include pandas nullable ints
    for col in df_copy.select_dtypes(include=int_dtypes).columns:
        n_unique = df_copy[col].nunique(dropna=True)
        if n_unique <= max_unique_values:
            df_copy[col] = df_copy[col].astype("category")
            changed.append((col, n_unique))

    if verbose and changed:
        print("Converted to 'category':", ", ".join(f"{c} (n_unique={k})" for c, k in changed))

    return df_copy
```

```python
#Define int_to_categorical
def int_to_categorical(df, max_unique_values=20, verbose=True):
    """Convert low-cardinality integer columns to 'category' dtype."""
    df_copy = df.copy()
    changed = []

    # include pandas nullable ints too ("Int32","Int64")
    int_dtypes = ["int", "int32", "int64", "Int32", "Int64"]

    for col in df_copy.select_dtypes(include=int_dtypes).columns:
        n_unique = df_copy[col].nunique(dropna=True)
        if n_unique <= max_unique_values:
            df_copy[col] = df_copy[col].astype("category")
            changed.append((col, n_unique))

    if verbose and changed:
        print("Converted to 'category':", ", ".join(f"{c} (n_unique={k})" for c, k in changed))

    return df_copy
```

```python
# Change relevant integer columns
df_new = int_to_categorical(df, max_unique_values=20)

# High-level metadata view
df_new.info(show_counts=True, memory_usage="deep")
```

```python
#create_metadata
import pandas as pd

def create_metadata(df):
    """
    Build a metadata table for each column in df.
    """
    meta = pd.DataFrame(index=df.columns)

    # core stats
    meta["dtype"]        = df.dtypes.astype(str)
    meta["n_unique"]     = df.nunique(dropna=True)
    meta["missing_sum"]  = df.isna().sum()
    meta["missing_pct"]  = (meta["missing_sum"] / len(df)).round(4)
    meta["non_null"]     = len(df) - meta["missing_sum"]

    # one example non-null value per column (handy for a quick glance)
    meta["example"] = df.apply(lambda s: s.dropna().iloc[0] if s.notna().any() else pd.NA)

    return meta
```

```python
import pandas as pd
from pandas.api.types import is_numeric_dtype, is_object_dtype, is_string_dtype

def create_metadata(df):
    meta = pd.DataFrame(index=df.columns)
    meta["dtype_obj"]   = df.dtypes
    meta["dtype"]       = meta["dtype_obj"].astype(str)
    meta["n_unique"]    = df.nunique(dropna=True)
    meta["missing_sum"] = df.isna().sum()
    meta["missing_pct"] = (meta["missing_sum"] / len(df)).round(4)
    meta["non_null"]    = len(df) - meta["missing_sum"]

    # Updated checks (no deprecation warnings):
    meta["is_numeric"] = meta["dtype_obj"].apply(is_numeric_dtype)
    meta["is_categorical"] = meta["dtype_obj"].apply(
        lambda dt: isinstance(dt, pd.CategoricalDtype) or is_object_dtype(dt) or is_string_dtype(dt)
    )

    meta["sample_values"] = df.apply(lambda col: col.dropna().unique()[:5].tolist())
    return meta
```
