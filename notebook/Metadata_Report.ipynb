```python
def clean_df(df, max_unique_values=20, verbose=True):
    df_copy = df.copy()
    changed = []

    # cast low-cardinality integer columns to 'category'
    int_dtypes = ["int", "int32", "int64", "Int32", "Int64"]  # include pandas nullable ints
    for col in df_copy.select_dtypes(include=int_dtypes).columns:
        n_unique = df_copy[col].nunique(dropna=True)
        if n_unique <= max_unique_values:
            df_copy[col] = df_copy[col].astype("category")
            changed.append((col, n_unique))

    if verbose and changed:
        print("Converted to 'category':", ", ".join(f"{c} (n_unique={k})" for c, k in changed))

    return df_copy
```

```python
#Define int_to_categorical
def int_to_categorical(df, max_unique_values=20, verbose=True):
    """Convert low-cardinality integer columns to 'category' dtype."""
    df_copy = df.copy()
    changed = []

    # include pandas nullable ints too ("Int32","Int64")
    int_dtypes = ["int", "int32", "int64", "Int32", "Int64"]

    for col in df_copy.select_dtypes(include=int_dtypes).columns:
        n_unique = df_copy[col].nunique(dropna=True)
        if n_unique <= max_unique_values:
            df_copy[col] = df_copy[col].astype("category")
            changed.append((col, n_unique))

    if verbose and changed:
        print("Converted to 'category':", ", ".join(f"{c} (n_unique={k})" for c, k in changed))

    return df_copy
```

```python
#create_metadata
import pandas as pd

def create_metadata(df):
    """
    Build a metadata table for each column in df.
    """
    meta = pd.DataFrame(index=df.columns)

    # core stats
    meta["dtype"]        = df.dtypes.astype(str)
    meta["n_unique"]     = df.nunique(dropna=True)
    meta["missing_sum"]  = df.isna().sum()
    meta["missing_pct"]  = (meta["missing_sum"] / len(df)).round(4)
    meta["non_null"]     = len(df) - meta["missing_sum"]

    # one example non-null value per column (handy for a quick glance)
    meta["example"] = df.apply(lambda s: s.dropna().iloc[0] if s.notna().any() else pd.NA)

    return meta
```

```python
import pandas as pd
from pandas.api.types import is_numeric_dtype, is_object_dtype, is_string_dtype

def create_metadata(df):
    meta = pd.DataFrame(index=df.columns)
    meta["dtype_obj"]   = df.dtypes
    meta["dtype"]       = meta["dtype_obj"].astype(str)
    meta["n_unique"]    = df.nunique(dropna=True)
    meta["missing_sum"] = df.isna().sum()
    meta["missing_pct"] = (meta["missing_sum"] / len(df)).round(4)
    meta["non_null"]    = len(df) - meta["missing_sum"]

    # Updated checks (no deprecation warnings):
    meta["is_numeric"] = meta["dtype_obj"].apply(is_numeric_dtype)
    meta["is_categorical"] = meta["dtype_obj"].apply(
        lambda dt: isinstance(dt, pd.CategoricalDtype) or is_object_dtype(dt) or is_string_dtype(dt)
    )

    meta["sample_values"] = df.apply(lambda col: col.dropna().unique()[:5].tolist())
    return meta
```

```python
#loop-based version of create_metadata
import pandas as pd
from pandas.api.types import is_numeric_dtype, is_object_dtype

def create_metadata_loop(df):
    """Build a metadata table for each column in df (loop version)."""
    rows = []
    n = len(df)

    for col in df.columns:
        s = df[col]
        dtype = s.dtype
        n_unique = s.nunique(dropna=True)
        missing_sum = int(s.isna().sum())
        missing_pct = round(missing_sum / n, 4) if n else 0.0
        non_null = n - missing_sum

        # example non-missing value (if any)
        if non_null:
            try:
                example = s.dropna().iloc[0]
            except Exception:
                example = pd.NA
        else:
            example = pd.NA

        rows.append({
            "column": col,
            "dtype": str(dtype),
            "n_unique": int(n_unique),
            "missing_sum": missing_sum,
            "missing_pct": missing_pct,
            "non_null": int(non_null),
            "example": example,
            # optional flags (uncomment if you want them)
            # "is_numeric": bool(is_numeric_dtype(dtype)),
            # "is_categorical": bool(is_object_dtype(dtype) or isinstance(dtype, pd.CategoricalDtype)),
        })

    meta = pd.DataFrame(rows).set_index("column")
    return meta
```

```python
import re
import pandas as pd
from pandas.api.types import is_numeric_dtype

def create_metadata_smart(
    df,
    max_unique_values=20,
    cat_name_patterns=(
        r"^attr_",                        # all attribute flags
        r"restaurantspricerange",         # e.g., attr_RestaurantsPriceRange2
        r"price_range", r"pricerange",    # common names
        r"pricelevel", r"price_tier"
    ),
    treat_binary_as_categorical=True,
    extras_top_values=5
):
    """
    Build a metadata table and *suggest* categorical vs numeric based on:
      - name patterns (attr_*, price range),
      - dtype (object/category),
      - low-cardinality integers,
      - binary 0/1 columns (optional).
    """
    meta = pd.DataFrame(index=df.columns)
    n = len(df)

    # Base stats
    meta["dtype_obj"]   = df.dtypes
    meta["dtype"]       = meta["dtype_obj"].astype(str)
    meta["n_unique"]    = df.nunique(dropna=True)
    meta["missing_sum"] = df.isna().sum()
    meta["missing_pct"] = (meta["missing_sum"] / n).round(4)
    meta["non_null"]    = n - meta["missing_sum"]

    # Helper: decide if a column name matches any categorical pattern (case-insensitive)
    def _name_looks_categorical(col):
        low = col.lower()
        return any(re.search(pat, low) for pat in cat_name_patterns)

    reasons = []
    suggested = []
    for col in df.columns:
        s = df[col]
        dt = s.dtype
        reason = None
        is_cat = False

        # 1) Name pattern wins (attr_*, price range, etc.)
        if _name_looks_categorical(col):
            is_cat, reason = True, "name_pattern"

        # 2) Object / string or already category
        elif str(dt) == "category" or s.dtype == "object":
            is_cat, reason = True, "object_or_category"

        # 3) Binary 0/1 (even if stored as int)
        elif treat_binary_as_categorical:
            vals = s.dropna().unique()
            # Treat bool dtype or {0,1}/{1,0} as categorical
            if str(dt) == "bool" or (len(vals) <= 2 and set(pd.Series(vals).dropna().astype(str)).issubset({"0","1"})):
                is_cat, reason = True, "binary"

        # 4) Low-cardinality integers (e.g., 1â€“4 price tier)
        if not is_cat and is_numeric_dtype(dt):
            nunq = int(s.nunique(dropna=True))
            if nunq <= max_unique_values:
                is_cat, reason = True, f"low_card_int(n_unique={nunq})"

        # Decide suggestion
        suggested.append("category" if is_cat else "numeric")
        reasons.append(reason if reason else "numeric_default")

    meta["suggested_dtype"] = suggested
    meta["categorical_reason"] = reasons

    # Sample/example values
    meta["sample_values"] = df.apply(lambda col: col.dropna().unique()[:5].tolist())

    # For categorical columns, add top-value snapshot (helps spot weird encodings)
    def _top_vals(col):
        vc = df[col].value_counts(dropna=True).head(extras_top_values)
        return [f"{idx}:{cnt}" for idx, cnt in vc.items()]

    meta["top_values"] = [
        _top_vals(c) if meta.loc[c, "suggested_dtype"] == "category" else []
        for c in df.columns
    ]

    return meta
```

```python
# Build metadata
meta = create_metadata_smart(df)        # or df_new
meta.head(20)

# Convert suggested categorical columns to 'category' dtype
cat_cols = meta.index[meta["suggested_dtype"] == "category"].tolist()
df_converted = df.copy()
for c in cat_cols:
    df_converted[c] = df_converted[c].astype("category")

# Verify
meta_converted = create_metadata_smart(df_converted)
meta_converted.loc[cat_cols, ["dtype", "suggested_dtype", "categorical_reason"]].head(10)
```

```python
import re
import pandas as pd
from pandas.api.types import is_numeric_dtype, is_object_dtype, is_string_dtype

ID_PATTERNS = (r"(^|_)id$", r"^business_id$", r"^user_id$", r"^review_id$")

def looks_like_id(col: str) -> bool:
    low = col.lower()
    return any(re.search(p, low) for p in ID_PATTERNS)

def create_metadata_smart(df, max_unique_values=20):
    meta = pd.DataFrame(index=df.columns)
    n = len(df)

    meta["dtype_obj"]   = df.dtypes
    meta["dtype"]       = meta["dtype_obj"].astype(str)
    meta["n_unique"]    = df.nunique(dropna=True)
    meta["missing_sum"] = df.isna().sum()
    meta["missing_pct"] = (meta["missing_sum"] / n).round(4)

    # suggestions
    suggested = []
    reason = []
    is_id = []

    for c in df.columns:
        s, dt = df[c], df[c].dtype

        if looks_like_id(c):
            suggested.append("identifier")
            reason.append("name_is_id")
            is_id.append(True)
            continue

        # categorical by type/name/low-card int
        if str(dt) == "category" or is_object_dtype(dt) or is_string_dtype(dt):
            suggested.append("category"); reason.append("object_or_category")
        elif is_numeric_dtype(dt) and s.nunique(dropna=True) <= max_unique_values:
            suggested.append("category"); reason.append(f"low_card_int(n={s.nunique(dropna=True)})")
        else:
            suggested.append("numeric");  reason.append("numeric_default")
        is_id.append(False)

    meta["suggested_dtype"]   = suggested
    meta["categorical_reason"] = reason
    meta["is_identifier"]     = is_id

    return meta
```

```python
# 1) Build metadata
meta = create_metadata_smart(df)

# 2) Decide what to convert
cat_cols = meta.index[(meta["suggested_dtype"] == "category") & (~meta["is_identifier"])]

# 3) Make a fixed copy:
df_fixed = df.copy()
# Cast IDs to string (optional but recommended)
id_cols = meta.index[meta["is_identifier"]]
for c in id_cols:
    df_fixed[c] = df_fixed[c].astype("string")
# Cast categorical (excluding IDs)
for c in cat_cols:
    df_fixed[c] = df_fixed[c].astype("category")

# 4) Rebuild metadata to verify dtypes
meta_fixed = create_metadata_smart(df_fixed)
meta_fixed.loc[list(id_cols) + list(cat_cols), ["dtype", "suggested_dtype", "is_identifier"]]
```
